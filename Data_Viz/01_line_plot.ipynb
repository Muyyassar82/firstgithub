## Adding Titles

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset   
phool = sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length" , y="sepal_width", data=phool)
plt.title("pholo ka plot")
plt.show()
```

## Adding Limits

```python
# Limit Axis
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset   
phool = sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length" , y="sepal_width", data=phool)
plt.title("pholo ka plot")
plt.xlim(2)
plt.ylim(1)

plt.show()
```

## Set Style
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
# Limit Axis
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset   
phool = sns.load_dataset("iris")
phool
#draw a line plot
sns.lineplot(x="sepal_length" , y="sepal_width", data=phool, color='red', linestyle='dotted', linewidth=2.5)
plt.title("pholo ka plot")
plt.xlim(2)
plt.ylim(1)

plt.show()
```

## Size of figure

```python
# Limit Axis
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset   
phool = sns.load_dataset("iris")
phool

#Change figure size

plt.figure(figsize=(10,6))

#draw a line plot
sns.lineplot(x="sepal_length" , y="sepal_width", data=phool, color='red', linestyle='dotted', linewidth=2.5)
plt.title("pholo ka plot")
plt.xlim(2)
plt.ylim(1)

plt.show()
```
